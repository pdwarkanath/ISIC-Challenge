## Load and display training images

```python
# Check a random image from the training set

def display_rand_img(img_loc, segmentation = False):
    while True:
        try:
            last_img_idx = 16072
            a = str(np.random.choice(last_img_idx))
            a = a.zfill(7)
            if segmentation:
                img_path = '{}/ISIC_{}_segmentation.png'.format(img_loc, a)
            else:
                img_path = '{}/ISIC_{}.jpg'.format(img_loc, a)
            img = plt.imread(img_path)
            print(img_path)
            plt.axis('off')
            if segmentation:
                plt.imshow(img, cmap = 'gray')
            else:
                plt.imshow(img)
            plt.show()
            break
        except FileNotFoundError:
            continue
```

```python
# Get a list of all image files in training set

X_train_filelist = glob.glob('../ISIC2018_Task1-2_Training_Input/*.jpg')
```

```python
# Make empty array to fill up with training image data of appropriate size

m_train_orig = len(X_train_filelist)
nH0, nW0, nC0 = (512, 512, 3)
```

```python
# Fill array with image in the right format
def make_img_arrays(filelist, nH0, nW0):

    lst = []
    for i, fname in enumerate(filelist):
        img = image.load_img(fname, target_size=(nH0, nW0))
        img_array = image.img_to_array(img, dtype = np.uint8)/255.
        img_array = np.expand_dims(img_array, axis = 0)
        lst.append(img_array)
        if i%100 == 0:
            print(fname)
    arr = np.vstack(lst)
    return arr
```

```python
X_train_orig = make_img_arrays(X_train_filelist[:100], nH0, nW0)
```

```python
def iou(A, B):
    # This function calculates the intersection over union of 2 matrices
    # A, B - matrix of size nW x nH with 0s and 1s in each position
    # 0 denotes dark spot and 1 denotes bright spot
    # intersection is the bright area common to both matrices
    # union is the total bright area 
    intersection = A*B
    union = np.minimum(A+B, 1)
    iou = np.sum(intersection)/np.sum(union)
    
    return iou
```
